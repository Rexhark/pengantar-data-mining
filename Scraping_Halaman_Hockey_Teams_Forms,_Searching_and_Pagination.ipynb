```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
teams = []

for page_num in range(6):
  url = "https://www.scrapethissite.com/pages/forms/?page_num=" + str(page_num + 1) + "&per_page=100"
  respons = requests.get(url)
  soup = BeautifulSoup(respons.content)
  teams.extend(soup.find_all(class_="team"))
```

```python
len(teams)
```

```python
name = []
year = []
wins = []
losses = []
ot_losses = []
win_pct = []
gf = []
ga = []
diff = []

for team in teams:
  name.append(team.find(class_="name").text.strip())
  year.append(team.find(class_="year").text.strip())
  wins.append(team.find(class_="wins").text.strip())
  losses.append(team.find(class_="losses").text.strip())
  ot_losses.append(team.find(class_="ot-losses").text.strip())
  win_pct.append(team.find(class_="pct").text.strip())
  gf.append(team.find(class_="gf").text.strip())
  ga.append(team.find(class_="ga").text.strip())
  diff.append(team.find(class_="diff").text.strip())
```

```python
FinalData = pd.DataFrame(
    {'name': name,
     'year':year,
     'wins':wins,
     'losses':losses,
     'ot_losses':ot_losses,
     'win_pct':win_pct,
     'gf':gf,
     'ga':ga,
     'diff':diff
    })

FinalData
```

```python
FinalData.to_csv('FinalData.csv')
```
